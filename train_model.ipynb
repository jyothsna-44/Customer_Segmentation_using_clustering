```python
import pandas as pd
import pickle
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# 1. Load Data
df = pd.read_excel('Mall_Customers.xlsx')

# 2. FEATURE SELECTION (Using Names instead of Numbers)
# Replace 'Annual Income (k$)' and 'Spending Score (1-100)' 
# with the EXACT names from your Excel header row.
X = df[['Annual Income (k$)', 'Spending Score (1-100)']] 

# 3. Data Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) # Now it won't crash!

# 4. Model Training
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
kmeans.fit(X_scaled)

# 5. Save Model and Scaler
pickle.dump(kmeans, open('kmeans_model.pkl', 'wb'))
pickle.dump(scaler, open('scaler.pkl', 'wb'))

print("Success: kmeans_model.pkl created without errors!")
```
